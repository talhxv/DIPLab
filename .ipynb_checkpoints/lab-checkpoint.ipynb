```python
import random

# Players
player, opponent = 'x', 'o'

# Counter variables for nodes visited and evaluated
nodes_visited = 0
nodes_evaluated = 0

# This function returns true if there are moves remaining on the board. It returns false if there are no moves left to play.
def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                return True
    return False

# This is the evaluation function as discussed in the previous article (http://goo.gl/sJgv68)
def evaluate(b):
    # Checking for Rows for X or O victory.
    for row in range(3):
        if b[row][0] == b[row][1] and b[row][1] == b[row][2]:
            if b[row][0] == player:
                return 10
            elif b[row][0] == opponent:
                return -10

    # Checking for Columns for X or O victory.
    for col in range(3):
        if b[0][col] == b[1][col] and b[1][col] == b[2][col]:
            if b[0][col] == player:
                return 10
            elif b[0][col] == opponent:
                return -10

    # Checking for Diagonals for X or O victory.
    if b[0][0] == b[1][1] and b[1][1] == b[2][2]:
        if b[0][0] == player:
            return 10
        elif b[0][0] == opponent:
            return -10

    if b[0][2] == b[1][1] and b[1][1] == b[2][0]:
        if b[0][2] == player:
            return 10
        elif b[0][2] == opponent:
            return -10

    # Else if none of them have won then return 0
    return 0

# This is the minimax function with alpha-beta pruning. It considers all the possible ways the game can go and returns the value of the board
def minimax(board, depth, isMax, alpha, beta):
    global nodes_visited, nodes_evaluated
    
    score = evaluate(board)

    # If Maximizer has won the game return his/her evaluated score
    if score == 10:
        return score

    # If Minimizer has won the game return his/her evaluated score
    if score == -10:
        return score

    # If there are no more moves and no winner then it is a tie
    if not isMovesLeft(board):
        return 0

    # If this maximizer's move
    if isMax:
        best = -1000

        # Traverse all cells
        for i in range(3):
            for j in range(3):
                # Check if cell is empty
                if board[i][j] == '_':
                    # Make the move
                    board[i][j] = player

                    # Increment nodes visited count
                    nodes_visited += 1

                    # Call minimax recursively and choose the maximum value
                    best = max(best, minimax(board, depth + 1, not isMax, alpha, beta))

                    # Undo the move
                    board[i][j] = '_'

                    # Update alpha
                    alpha = max(alpha, best)

                    # Alpha-beta pruning
                    if beta <= alpha:
                        break
        return best

    # If this minimizer's move
    else:
        best = 1000

        # Traverse all cells
        for i in range(3):
            for j in range(3):
                # Check if cell is empty
                if board[i][j] == '_':
                    # Make the move
                    board[i][j] = opponent

                    # Increment nodes visited count
                    nodes_visited += 1

                    # Call minimax recursively and choose the minimum value
                    best = min(best, minimax(board, depth + 1, not isMax, alpha, beta))

                    # Undo the move
                    board[i][j] = '_'

                    # Update beta
                    beta = min(beta, best)

                    # Alpha-beta pruning
                    if beta <= alpha:
                        break
        return best

# This will return the best possible move for the player
def findBestMove(board):
    global nodes_visited, nodes_evaluated
    bestVal = -1000
    bestMove = (-1, -1)
    searchSpace = 0  # Counter for search space

    # Traverse all cells, evaluate minimax function for all empty cells. And return the cell with optimal value.
    for i in range(3):
        for j in range(3):
            # Check if cell is empty
            if board[i][j] == '_':
                # Make the move
                board[i][j] = player

                # Increment search space counter
                searchSpace += 1

                # Compute evaluation function for this move.
                moveVal = minimax(board, 0, False, -1000, 1000)

                # Undo the move
                board[i][j] = '_'

                # Increment nodes evaluated count
                nodes_evaluated += 1

                # If the value of the current move is more than the best value, then update best
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal

    print("The value of the best Move is:", bestVal)
    print("Search Space:", searchSpace)
    print("Nodes Visited:", nodes_visited)
    print("Nodes Evaluated:", nodes_evaluated)
    print("Generated Board:")
    for row in board:
        print(" ".join(row))
    print("The Optimal Move is:")
    print("ROW:", bestMove[0], " COL:", bestMove[1])
    print()
    return bestMove

# Generate a random board
def generateRandomBoard():
    board = [['_', '_', '_'], ['_', '_', '_'], ['_', '_', '_']]
    for i in range(3):
        for j in range(3):
            board[i][j] = random.choice(['x', 'o', '_'])
    return board

# Driver code
random.seed()  # Optional: Use a fixed seed to get reproducible results
board = generateRandomBoard()

# Display the random board
print("Generated Board:")
for row in board:
    print(" ".join(row))

bestMove = findBestMove(board)
```

```python
import random

# Players
player, opponent = 'x', 'o'

# Function to check if there are moves remaining on the board
def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                return True
    return False

# Function to evaluate the board state
def evaluate(board):
    # Evaluation logic omitted for simplicity
    return 0

# Alpha-Beta Pruning algorithm
def alphaBeta(board, depth, alpha, beta, isMax):
    # Base cases omitted for simplicity
    if depth == 0 or not isMovesLeft(board):
        return evaluate(board)

    # Randomly shuffle the order of moves
    moves = [(i, j) for i in range(3) for j in range(3) if board[i][j] == '_']
    random.shuffle(moves)

    if isMax:
        value = float('-inf')
        for move in moves:
            board[move[0]][move[1]] = player
            value = max(value, alphaBeta(board, depth - 1, alpha, beta, not isMax))
            alpha = max(alpha, value)
            board[move[0]][move[1]] = '_'
            if beta <= alpha:
                break
        return value
    else:
        value = float('inf')
        for move in moves:
            board[move[0]][move[1]] = opponent
            value = min(value, alphaBeta(board, depth - 1, alpha, beta, not isMax))
            beta = min(beta, value)
            board[move[0]][move[1]] = '_'
            if beta <= alpha:
                break
        return value

# Function to make a move using Alpha-Beta Pruning
def makeAlphaBetaMove(board):
    # Initialize alpha and beta values
    alpha = float('-inf')
    beta = float('inf')
    best_move = None

    # Randomly shuffle the order of moves
    moves = [(i, j) for i in range(3) for j in range(3) if board[i][j] == '_']
    random.shuffle(moves)

    # Explore each possible move
    for move in moves:
        board[move[0]][move[1]] = player
        move_value = alphaBeta(board, 3, alpha, beta, False)
        if move_value > alpha:
            alpha = move_value
            best_move = move
        board[move[0]][move[1]] = '_'

    return best_move

# Function to play the game
def playGame():
    # Initialize the board
    board = [['_' for _ in range(3)] for _ in range(3)]

    # Main game loop
    while isMovesLeft(board):
        # Weak player's turn
        weak_move = makeAlphaBetaMove(board)
        if weak_move:
            board[weak_move[0]][weak_move[1]] = opponent

        # Display the board
        print("Current Board:")
        for row in board:
            print(" ".join(row))
        print()

        # Check if weak player won
        if evaluate(board) == -10:
            print("Weak player wins!")
            return

        # Player's turn
        player_move = makeAlphaBetaMove(board)
        if player_move:
            board[player_move[0]][player_move[1]] = player

        # Display the board
        print("Current Board:")
        for row in board:
            print(" ".join(row))
        print()

        # Check if player won
        if evaluate(board) == 10:
            print("Player wins!")
            return

    print("Game ended in a tie!")

playGame()
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import LabelEncoder

class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions

    def _predict(self, x):
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = Counter(k_nearest_labels).most_common(1) 
        return most_common[0][0]

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

df = pd.read_csv('sdata.csv')

le = LabelEncoder()
df[' Category'] = le.fit_transform(df[' Category'])

features = df[['Feature1', ' Feature2', ' Feature3', ' Category']].values
labels = df[' Label'].values
knn = KNN(k=5)
knn.fit(features, labels)
new_point = np.array([9.8, 5.7, 3.7, 1])
prediction = knn.predict([new_point])

print("The label assigned to the point (9.8, 5.7, 3.7, B) is:", prediction[0])
```
